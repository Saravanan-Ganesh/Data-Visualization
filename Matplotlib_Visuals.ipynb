```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(12,20)
y=np.arange(1,9)
```

```python
a=np.arange(10,20)
b=np.arange(60,70)
```

```python
plt.plot(x,y)
plt.scatter(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python

plt.scatter(a,b,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
y=x*x
```

```python
#plt.scatter(x,y,marker='+',linewidths=2)
plt.plot(x,y,marker="+",linestyle='-',linewidth=2,c='r',markersize=10)
```

```python
plt.subplot(3,2,1)
plt.plot(x,y,marker="+")
plt.subplot(3,2,2)
plt.scatter(x,y,c='r')
plt.subplot(3,2,3)
plt.plot(a,b,marker="h")
plt.subplot(3,2,4)
plt.scatter(a,b,marker=">")
plt.subplot(3,2,5)
plt.plot(x,y,marker="D")
plt.subplot(3,2,6)
plt.scatter(x,y,marker="o")
```

```python
plt.subplot(1,4,1)
plt.plot(x,y,marker="+")
plt.subplot(1,4,4)
plt.plot(x,y,marker="*")
plt.subplot(1,4,3)
plt.plot(x,y,marker="o")
plt.subplot(1,4,2)
plt.plot(x,y,marker="o")
```
